```python
# first we will split Schedule for each class
import json 
import os

file_path = r'data_final\emploi-temps\emploie.json' 
with open(file_path, 'r', encoding='utf-8') as file:
    schedule_data = json.load(file)
output_path =  r'data_final\emploi-temps'
for item, content in schedule_data.items():
    print(item)
    output_file = os.path.join(output_path, f'{item.lower()}.json')
    print('output_file', output_file)
    with open(output_file, 'w', encoding='utf-8') as file:
        json.dump(content, file, ensure_ascii=False, indent=4)
    print(f'{item} done')
```

```python
folder_data = 'data_final'
for root, dirs, files in os.walk(folder_data):
    for file in files:
        if file.endswith('.txt'):
            file_path = os.path.join(root, file)
            with open(file_path, 'r', encoding='utf-8') as infile:
                data = infile.readlines()
            
            # Remove empty lines
            cleaned_data = [line for line in data if line.strip()]

            # Write cleaned data back to the file
            with open(file_path, 'w', encoding='utf-8') as outfile:
                outfile.writelines(cleaned_data)
            print(f'File {file} cleaned')
```

```python
folder_path = r'C:\Users\user\Desktop\ci2\Projects\ChatBot ENSA\data\data_final\emploi-temps'
for file in os.listdir(folder_path):
    if file.endswith(".json"):  # Ensure only JSON files are processed
        with open(os.path.join(folder_path, file), 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        chunk = f'{file} : '  # Store file name as metadata
        for day, schedule in data.items():
            chunk += f' | {day} : '  # Add day to chunk
            
            for entry in schedule:
                entry_text = " | ".join([f"{k}: {v}" for k, v in entry.items()])
                chunk += f' {entry_text} ;'  # Append schedule details
    print(chunk)
    break
```
