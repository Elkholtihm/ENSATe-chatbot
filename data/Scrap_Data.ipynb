```python
# Step 1: Extract all links from each sitemap
def get_all_links(sitemap_urls):
    all_links = []
    for sitemap_url in sitemap_urls:
        response = requests.get(sitemap_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "xml")
            links = [loc.text for loc in soup.find_all("loc")]
            all_links.extend(links)
        else:
            print(f"Failed to fetch {sitemap_url}")
    return all_links
```

```python
links = lines.strip().split('\n')
links = [x for x in links if x!='']
formationL = links[1:11]
```

```python
def scrap(url, tags):
    try:
        response = requests.get(url)
        response.encoding = response.apparent_encoding
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            elements = soup.find_all(tags)
            return "\n".join(element.get_text(strip=True) for element in elements)
        else: 
            print(f"Error: Status code {response.status_code}")
            return None
    except requests.exceptions.RequestException as e: 
        print(f"Request error: {e}")
        return None
```

```python
# Scarap fomration data
for url in formationL:
    result = scrap(url, ['tr', 'strong', 'p'])
    
    filename = url.split('/')[-2]
    # save data 
    with open(f'{filename}.txt', 'w', encoding='utf-8') as f:
        f.write(result)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Define the base URL for the website
base_url = "https://ensa-tetouan.ac.ma/"

DecLinks = {}

# Initialize Selenium
driver = webdriver.Chrome()
driver.get(base_url)


# Wait for the "Formations" menu to be clickable
try:
    formations_menu = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.LINK_TEXT, "Formations"))
    )
    formations_menu.click()

    # Wait briefly for sub-items to load
    time.sleep(2)

    # Parse page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Find sub-items under "Formations"
    sub_items = soup.find_all('a', class_="elementor-sub-item")
    for item in sub_items:
        title = item.text
        link = item['href']
        DecLinks[title] = link
        

finally:
    # Close the browser
    driver.quit()
```

```python
# Helper function to fetch and parse a page
def GetData(url, tags, cl=None, retries=3, timeout=10):
    for attempt in range(retries):
        response = requests.get(url, timeout=timeout)
        response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')

    main_text = []
    for paragraph in soup.find_all(tags, class_=cl):
        main_text.append(paragraph.get_text())
    return main_text

#['td', 'p', 'h1', 'h2', 'u1', 'li']
```
